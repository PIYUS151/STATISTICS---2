# THEORY QUESTIONS

1. What is hypothesis testing in statistics?

->Hypothesis testing is a statistical method used to decide whether there is enough evidence in a sample to infer that a condition holds for the population.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

->Null Hypothesis (H₀): Assumes no effect or difference (status quo).

Alternative Hypothesis (H₁ or Ha): Assumes a significant effect or difference.

Example:
H₀: μ = 100 (no change)
H₁: μ ≠ 100 (there is a change)

 3. What is the significance level in hypothesis testing, and why is it important?

->The significance level (α) is the probability of rejecting the null hypothesis when it's true (Type I error). Common values are 0.05 or 0.01.

It defines the threshold for statistical significance.

 4. What does a P-value represent in hypothesis testing?

->The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming H₀ is true.

 5. How do you interpret the P-value in hypothesis testing?

->If P ≤ α → Reject H₀ (evidence is significant)

If P > α → Fail to reject H₀ (evidence is not significant)

 6. What are Type 1 and Type 2 errors in hypothesis testing?

->Type I Error (False Positive): Rejecting H₀ when it's true (controlled by α)

Type II Error (False Negative): Failing to reject H₀ when it's false (denoted by β)

 7. What is the difference between a one-tailed and a two-tailed test?

->One-tailed test: Tests deviation in one direction (e.g., H₁: μ > μ₀)

Two-tailed test: Tests deviation in both directions (e.g., H₁: μ ≠ μ₀)

 8. What is the Z-test, and when is it used?

->A Z-test is used when:

Sample size is large (n > 30)

Population variance is known
It checks if sample mean significantly differs from population mean.

 10. What is the T-distribution, and when should it be used instead of the normal distribution?

->Use the T-distribution when:

Sample size is small (n < 30)

Population standard deviation is unknown


11. What is the difference between a Z-test and a T-test?

->Test	When to Use

 12. What is the T-test, and how is it used in hypothesis testing?

->The T-test checks if the means of one or two groups differ significantly.

One-sample T-test: Compare sample mean to population mean

Two-sample T-test: Compare means of two independent groups

Paired T-test: Compare before/after in the same group

 13. What is the relationship between Z-test and T-test?
->Both test for differences in means:

T-test approximates Z-test as sample size increases

As n → ∞, T-distribution ≈ Normal distribution

 14. What is a confidence interval, and how is it used to interpret statistical results?

->A confidence interval (CI) estimates a range where the true population parameter lies with a given level of confidence (e.g., 95%).

If a 95% CI for μ is [48, 52], we are 95% confident the true mean lies in this range.

->The margin of error is the maximum expected difference between the sample and population value.
Wider margin → wider confidence interval → more uncertainty

 17. What is the Chi-square distribution, and when is it used?

T->he Chi-square distribution is used in tests involving categorical variables:

Goodness of fit

Test of independence

 18. What is the Chi-square goodness of fit test, and how is it applied?
->It checks whether observed frequencies match expected frequencies in a categorical dataset.
Used when testing if data follows a specific distribution.

 19. What is the F-distribution, and when is it used in hypothesis testing?

->The F-distribution is used to compare variances between groups.
It’s used in ANOVA and F-tests.

 20. What is an ANOVA test, and what are its assumptions?

->ANOVA (Analysis of Variance) compares means of 3+ groups.
Assumptions:

Independence of observations

Normally distributed groups

Equal variances (homogeneity of variance)

 21. What are the different types of ANOVA tests?

->One-way ANOVA: One independent variable

Two-way ANOVA: Two independent variables

Repeated Measures ANOVA: Same group tested multiple times

 22. What is the F-test, and how does it relate to hypothesis testing?

->The F-test compares two variances. It forms the basis of ANOVA.
If the F-statistic is large enough, it suggests significant variance differences between groups.

```python
#1 Perform a Z-test for a Sample Mean

from scipy import stats

# Known population parameters
population_mean = 100
population_std = 15
sample = [102, 98, 100, 105, 97, 103, 99]

sample_mean = sum(sample) / len(sample)
n = len(sample)
z_score = (sample_mean - population_mean) / (population_std / (n ** 0.5))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
print("Reject H0" if p_value < 0.05 else "Fail to reject H0")
```

```python
#2 Simulate Data and Perform Hypothesis Testing

import numpy as np
from scipy.stats import ttest_1samp

np.random.seed(0)
data = np.random.normal(loc=52, scale=10, size=100)
t_stat, p_value = ttest_1samp(data, popmean=50)

print("T-statistic:", t_stat)
print("P-value:", p_value)
print("Reject H0" if p_value < 0.05 else "Fail to reject H0")
```

```python
#3 One-Sample Z-test

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = np.array([105, 98, 100, 102, 101])
z, p = one_sample_z_test(sample, 100, 15)
print("Z:", z, "| P-value:", p)
```

```python
#4 Two-Tailed Z-test with Visualization

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 100
sigma = 15
sample = np.array([105, 98, 100, 102, 101])
sample_mean = np.mean(sample)
n = len(sample)
z_score = (sample_mean - mu) / (sigma / np.sqrt(n))
alpha = 0.05

# Plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < -norm.ppf(1 - alpha/2)), color='red', alpha=0.3, label="Rejection Region")
plt.fill_between(x, y, where=(x > norm.ppf(1 - alpha/2)), color='red', alpha=0.3)
plt.axvline(z_score, color='blue', label=f"Z-score: {z_score:.2f}")
plt.title("Two-Tailed Z-Test")
plt.legend()
plt.show()
```

```python
#5 Function to Visualize Type 1 and Type 2 Errors

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    x = np.linspace(90, 115, 1000)
    h0 = norm(mu0, se)
    h1 = norm(mu1, se)

    crit_val = norm.ppf(1 - alpha, loc=mu0, scale=se)

    plt.plot(x, h0.pdf(x), label='H0 Distribution')
    plt.plot(x, h1.pdf(x), label='H1 Distribution')
    plt.axvline(crit_val, color='red', linestyle='--', label='Critical Value')

    plt.fill_between(x, h0.pdf(x), where=(x >= crit_val), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, h1.pdf(x), where=(x < crit_val), color='blue', alpha=0.3, label='Type II Error (β)')

    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.show()

visualize_errors()
```

```python
#6  Independent T-test

import numpy as np
from scipy.stats import ttest_ind

group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
t_stat, p_val = ttest_ind(group1, group2)

print("T-statistic:", t_stat)
print("P-value:", p_val)
print("Reject H0" if p_val < 0.05 else "Fail to reject H0")
```

```python
#7 Paired Sample T-test with Visualization

import seaborn as sns
import pandas as pd
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

before = np.random.normal(50, 5, 30)
after = before + np.random.normal(2, 2, 30)

t_stat, p_val = ttest_rel(before, after)
print("T-stat:", t_stat)
print("P-value:", p_val)

df = pd.DataFrame({'Before': before, 'After': after})
sns.boxplot(data=df)
plt.title("Paired T-Test: Before vs After")
plt.show()
```

```python
#9 Function to Calculate Confidence Interval

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    sem = np.std(sample, ddof=1) / np.sqrt(n)
    h = sem * t.ppf((1 + confidence) / 2, df=n-1)
    return (mean - h, mean + h)

sample = np.random.normal(100, 10, 30)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

```python
#10

import numpy as np
from scipy.stats import norm

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    z = norm.ppf(1 - (1 - confidence)/2)
    moe = z * std_err
    return moe

data = np.random.normal(50, 10, 100)
print("Margin of Error:", margin_of_error(data))
```

```python
#11 Bayesian Inference with Bayes’ Theorem

def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prior_A * prob_B_given_A) / prob_B

# Example: Disease test
P_D = 0.01
P_Pos_given_D = 0.99
P_Pos = 0.01 * 0.99 + 0.99 * 0.05

posterior = bayes_theorem(P_D, P_Pos_given_D, P_Pos)
print("P(Disease | Positive Test):", posterior)
```

```python
#16  F-test for Comparing Variances

from scipy.stats import f

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(50, 15, 100)

f_stat = np.var(data1, ddof=1) / np.var(data2, ddof=1)
df1, df2 = len(data1)-1, len(data2)-1
p_val = 1 - f.cdf(f_stat, df1, df2)

print("F-statistic:", f_stat, "P-value:", p_val)
```

```python
#17  ANOVA Test for Multiple Groups

from scipy.stats import f_oneway

group1 = np.random.normal(60, 10, 30)
group2 = np.random.normal(65, 10, 30)
group3 = np.random.normal(70, 10, 30)

f_stat, p_val = f_oneway(group1, group2, group3)
print("F-stat:", f_stat, "P-value:", p_val)
```

```python
#18  One-Way ANOVA + Plot

import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'value': np.concatenate([group1, group2, group3]),
    'group': ['G1']*30 + ['G2']*30 + ['G3']*30
})

sns.boxplot(x='group', y='value', data=df)
plt.title("One-way ANOVA: Group Comparison")
plt.show()
```

```python
#19  Check ANOVA Assumptions

import scipy.stats as stats

def check_anova_assumptions(*groups):
    # Normality
    for i, g in enumerate(groups):
        print(f"Group {i+1} Normality (Shapiro):", stats.shapiro(g))

    # Homogeneity of variance
    print("Equal Variance (Levene’s test):", stats.levene(*groups))

check_anova_assumptions(group1, group2, group3)
```

```python
#20  Two-Way ANOVA + Visualization

import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulated data
df = pd.DataFrame({
    'score': np.random.normal(70, 5, 90),
    'factor1': ['A']*30 + ['B']*30 + ['C']*30,
    'factor2': ['X']*15 + ['Y']*15 + ['X']*15 + ['Y']*15 + ['X']*15 + ['Y']*15
})

model = ols('score ~ C(factor1) * C(factor2)', data=df).fit()
anova_result = sm.stats.anova_lm(model, typ=2)
print(anova_result)

sns.boxplot(x='factor1', y='score', hue='factor2', data=df)
plt.title("Two-Way ANOVA Visualization")
plt.show()
```
